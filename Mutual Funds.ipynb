```python
import pandas as pd
import numpy as np
```

```python
file_path = 'Historical Returns.xlsx'
df = pd.read_excel(file_path, skiprows=1)
df.columns = ["Scheme_Name", "Plan", "Category", "Crisil_Rating", "AuM", "1W", "1M", "3M", "6M", "YTD", "1Y", "2Y", "3Y", "5Y", "10Y"]
```

```python
df[['3Y', '5Y', '10Y']] = df[['3Y', '5Y', '10Y']].replace("-", np.nan)
df[['3Y', '5Y', '10Y']] = df[['3Y', '5Y', '10Y']].infer_objects(copy=False).astype(float)
```

```python
df['3Y'] = df['3Y'].fillna(df['3Y'].median())
df['5Y'] = df['5Y'].fillna(df['5Y'].median())
df['10Y'] = df['10Y'].fillna(df['10Y'].median())
```

```python
print(df[['3Y', '5Y', '10Y']].tail())
```

```python
# Function to get top performers for each category
def get_top_performers(df, category, top_n=5):
    # Filter the data for the selected category
    category_df = df[df['Category'] == category]
    
    # Sorting by 3Y returns (you can also sort by 1Y, 5Y, etc., based on user preference)
    top_performers = category_df.sort_values(by='3Y', ascending=False).head(top_n)
    
    return top_performers[['Scheme_Name', 'Category', 'Crisil_Rating', 'AuM', '1Y', '3Y', '5Y']]

```
